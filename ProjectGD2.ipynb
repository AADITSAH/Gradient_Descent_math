```python
def step_grad(x_train,y_train,lr,m):
    l=len(m)
    M=len(x_train)
    m_slope=np.array(m)
    x=[]
    for i in range(M):
        x=np.append(x_train[i],[1])
        x=np.array(x)
        s=(m*x).sum()
        for j in range(len(x)):
            m_slope[j]+=(-2/M)*((y_train[i]-s))*x[j]
    new_m=m-lr*m_slope
#     print(new_m)
    return new_m    
```

```python
def gd(x_train,y_train,learning_rate,num_iteration):
    n=len(x_train[1,:])
    array_m_c=np.zeros(n+1)
    for i in range(num_iteration):
        array_m_c=step_grad(x_train,y_train,learning_rate,array_m_c)
        print(i," cost ",cost(x_train,y_train,array_m_c))
    return array_m_c
```

```python
def cost(x_train,y_train, m):
    total_cost = 0
    M = len(x_train)
    for i in range(M):
        x=np.append(x_train[i],[1])
        x=np.array(x)
        s=(m*x).sum()
        
        total_cost += (1/M)*((y_train[i]-s)**2)
    return total_cost
```
